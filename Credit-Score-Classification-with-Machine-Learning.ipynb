```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error
from sklearn.preprocessing import StandardScaler
import numpy as np

credit_data = credit_data1.dropna(subset=['Credit_Score', 'Annual_Income', 'Monthly_Inhand_Salary', 
                                         'Num_Bank_Accounts', 'Num_Credit_Card', 'Interest_Rate',
                                         'Num_of_Loan', 'Delay_from_due_date', 'Num_of_Delayed_Payment',
                                         'Credit_Mix', 'Outstanding_Debt', 'Credit_Utilization_Ratio',
                                         'Credit_History_Age', 'Total_EMI_per_month',
                                         'Amount_invested_monthly', 'Monthly_Balance'])

credit_data['Credit_Score'] = credit_data['Credit_Score'].map({'Good':1,
                                                               'Poor':0,
                                                               'Standard':2})
credit_data['Credit_Mix'] = credit_data['Credit_Mix'].map({'Good':1,
                                                            'Bad':0,
                                                         'Standard':2})

x = np.array(credit_data[['Annual_Income', 'Monthly_Inhand_Salary', 'Num_Bank_Accounts',
                          'Num_Credit_Card', 'Interest_Rate', 'Num_of_Loan', 'Delay_from_due_date',
                          'Num_of_Delayed_Payment', 'Credit_Mix', 'Outstanding_Debt',
                          'Credit_Utilization_Ratio', 'Credit_History_Age', 'Total_EMI_per_month',
                          'Amount_invested_monthly', 'Monthly_Balance']])

y = np.array(credit_data['Credit_Score'])


model = LinearRegression() 

x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.3, random_state=42)


Standardscaler = StandardScaler()

x_train = Standardscaler.fit_transform(x_train)
x_test = Standardscaler.transform(x_test)

model.fit(x_train,y_train)

y_pred = model.predict(x_test)
#y_pred = model.predict(y_test)


mse = mean_squared_error(y_test,y_pred)
mae = mean_absolute_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)

print('mean_absolute_error:{:.4f}'.format(mse))
print('mean_squared_error:{:.4f}'.format(mae))
print('r2_score:{:.4f}'.format(r2))
```

logistic regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.preprocessing import StandardScaler

credit_data = credit_data1.dropna(subset=['Credit_Score', 'Annual_Income', 'Monthly_Inhand_Salary', 
                                         'Num_Bank_Accounts', 'Num_Credit_Card', 'Interest_Rate',
                                         'Num_of_Loan', 'Delay_from_due_date', 'Num_of_Delayed_Payment',
                                         'Credit_Mix', 'Outstanding_Debt', 'Credit_Utilization_Ratio',
                                         'Credit_History_Age', 'Total_EMI_per_month',
                                         'Amount_invested_monthly', 'Monthly_Balance'])

credit_data['Credit_Score'] = credit_data['Credit_Score'].map({'Good':1,
                                                               'Poor':0,
                                                               'Standard':2})
credit_data['Credit_Mix'] = credit_data['Credit_Mix'].map({'Good':1,
                                                            'Bad':0,
                                                         'Standard':2})

x = np.array(credit_data[['Annual_Income', 'Monthly_Inhand_Salary', 'Num_Bank_Accounts',
                          'Num_Credit_Card', 'Interest_Rate', 'Num_of_Loan', 'Delay_from_due_date',
                          'Num_of_Delayed_Payment', 'Credit_Mix', 'Outstanding_Debt',
                          'Credit_Utilization_Ratio', 'Credit_History_Age', 'Total_EMI_per_month',
                          'Amount_invested_monthly', 'Monthly_Balance','Credit_Score']])

y = np.array(credit_data['Credit_Score'])


x_train_loge, x_test_loge, y_train_loge, y_test_loge = train_test_split(x, y, test_size=0.3, random_state=42)

Standardscaler = StandardScaler()
x_train_loge = Standardscaler.fit_transform(x_train_loge)
x_test_loge = Standardscaler.transform(x_test_loge)

model = LogisticRegression()
model.fit(x_train_loge, y_train_loge)

y_pred_loge = model.predict(x_test_loge)

accuracyscore = accuracy_score(y_test_loge, y_pred_loge)
classificationreport = classification_report(y_test_loge, y_pred_loge)
confusionmatrix = confusion_matrix(y_test_loge, y_pred_loge)

print('accuracyscore of logistic regression: {:.4f}'.format(accuracyscore))
print('classification_report: \n',classificationreport)
print('confusionmatrix: \n',confusionmatrix)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,10))
plt.subplot(2,1,1)
plt.scatter(y_test_loge, y_pred_loge, color = 'blue',alpha=0.3)
plt.plot([min(y_test_loge),max(y_test_loge)],([min(y_pred_loge),max(y_pred_loge)]),color = 'red')
plt.title('actual values vs predicted values')
plt.xlabel('actual values')
plt.ylabel('predicted values')
plt.show()

plt.figure(figsize=(10,10))
plt.subplot(2,1,2)
sns.heatmap(confusionmatrix,cmap='Blues',annot=True,cbar=True,fmt='.1f')
plt.title('Confusion Matrix Heatmap')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,mean_squared_error
from sklearn.model_selection import train_test_split

credit_data = credit_data1.dropna(subset=['Credit_Score', 'Annual_Income', 'Monthly_Inhand_Salary', 
                                         'Num_Bank_Accounts', 'Num_Credit_Card', 'Interest_Rate',
                                         'Num_of_Loan', 'Delay_from_due_date', 'Num_of_Delayed_Payment',
                                         'Credit_Mix', 'Outstanding_Debt', 'Credit_Utilization_Ratio',
                                         'Credit_History_Age', 'Total_EMI_per_month',
                                         'Amount_invested_monthly', 'Monthly_Balance'])

credit_data['Credit_Score'] = credit_data['Credit_Score'].map({'Good':1,
                                                               'Poor':0,
                                                               'Standard':2})
credit_data['Credit_Mix'] = credit_data['Credit_Mix'].map({'Good':1,
                                                            'Bad':0,
                                                         'Standard':2})

x = np.array(credit_data[['Annual_Income', 'Monthly_Inhand_Salary', 'Num_Bank_Accounts',
                          'Num_Credit_Card', 'Interest_Rate', 'Num_of_Loan', 'Delay_from_due_date',
                          'Num_of_Delayed_Payment', 'Credit_Mix', 'Outstanding_Debt',
                          'Credit_Utilization_Ratio', 'Credit_History_Age', 'Total_EMI_per_month',
                          'Amount_invested_monthly', 'Monthly_Balance','Credit_Score']])

y = np.array(credit_data['Credit_Score'])

x_train_svc, x_test_svc, y_train_svc, y_test_svc = train_test_split(x, y, test_size=0.3, random_state=42)

Standardscaler = StandardScaler()
x_train_svc = Standardscaler.fit_transform(x_train_svc)
x_test_svc = Standardscaler.transform(x_test_svc)

model = SVC(kernel='poly', random_state=42)
model.fit(x_train_svc,y_train_svc)

y_pred_svc = model.predict(x_test_svc)

accuracyscore_svc = accuracy_score(y_test_svc,y_pred_svc)
classificationreport_svc = classification_report(y_test_svc,y_pred_svc)
confusionmatrix_svc = confusion_matrix(y_test_svc,y_pred_svc)

mean_squared_error_svc = mean_squared_error(y_test_svc,y_pred_svc)

print('mean_squared_error_svc: {:.4f}'.format(mean_squared_error_svc))

print('accuracyscore_svc: {:.4f}'.format(accuracyscore_svc))
print('classificationreport_svc: \n',classificationreport_svc)
print('confusionmatrix_svc: \n',confusionmatrix_svc)


```

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler

credit_data = credit_data1.dropna(subset=['Credit_Score', 'Annual_Income', 'Monthly_Inhand_Salary', 
                                         'Num_Bank_Accounts', 'Num_Credit_Card', 'Interest_Rate',
                                         'Num_of_Loan', 'Delay_from_due_date', 'Num_of_Delayed_Payment',
                                         'Credit_Mix', 'Outstanding_Debt', 'Credit_Utilization_Ratio',
                                         'Credit_History_Age', 'Total_EMI_per_month',
                                         'Amount_invested_monthly', 'Monthly_Balance'])

credit_data['Credit_Score'] = credit_data['Credit_Score'].map({'Good':1,
                                                               'Poor':0,
                                                               'Standard':2})
credit_data['Credit_Mix'] = credit_data['Credit_Mix'].map({'Good':1,
                                                            'Bad':0,
                                                         'Standard':2})

x = np.array(credit_data[['Annual_Income', 'Monthly_Inhand_Salary', 'Num_Bank_Accounts',
                          'Num_Credit_Card', 'Interest_Rate', 'Num_of_Loan', 'Delay_from_due_date',
                          'Num_of_Delayed_Payment', 'Credit_Mix', 'Outstanding_Debt',
                          'Credit_Utilization_Ratio', 'Credit_History_Age', 'Total_EMI_per_month',
                          'Amount_invested_monthly', 'Monthly_Balance']])

y = np.array(credit_data['Credit_Score'])

x_train_dt, x_test_dt, y_train_dt, y_test_dt = train_test_split(x, y, train_size=0.7, random_state=42)

tree = DecisionTreeClassifier(max_depth=2, random_state=42)

Standardscaler = StandardScaler()
x_train_dt = Standardscaler.fit_transform(x_train_dt)
x_test_dt = Standardscaler.transform(x_test_dt)

tree.fit(x_train_dt,y_train_dt)

y_pred_dt = tree.predict(x_test_dt)

accuracyscore_dt = accuracy_score(y_test_dt,y_pred_dt)
classificationreport_dt = classification_report(y_test_dt, y_pred_dt)
confusionmatrix_dt = confusion_matrix(y_test_dt,y_pred_dt)

print('accuracyscore_dt: {:.4f}'.format(accuracyscore_dt))
print('classificationreport_dt: \n',classificationreport_dt)
print('confusionmatrix_dt: \n',confusionmatrix_dt)

print('decisiontreeclassifier: \n',export_text(tree, feature_names=['Annual_Income', 'Monthly_Inhand_Salary', 'Num_Bank_Accounts',
                          'Num_Credit_Card', 'Interest_Rate', 'Num_of_Loan', 'Delay_from_due_date',
                          'Num_of_Delayed_Payment', 'Credit_Mix', 'Outstanding_Debt',
                          'Credit_Utilization_Ratio', 'Credit_History_Age', 'Total_EMI_per_month',
                          'Amount_invested_monthly', 'Monthly_Balance']))

```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

credit_data = credit_data1.dropna(subset=['Credit_Score', 'Annual_Income', 'Monthly_Inhand_Salary', 
                                         'Num_Bank_Accounts', 'Num_Credit_Card', 'Interest_Rate',
                                         'Num_of_Loan', 'Delay_from_due_date', 'Num_of_Delayed_Payment',
                                         'Credit_Mix', 'Outstanding_Debt', 'Credit_Utilization_Ratio',
                                         'Credit_History_Age', 'Total_EMI_per_month',
                                         'Amount_invested_monthly', 'Monthly_Balance'])

credit_data['Credit_Score'] = credit_data['Credit_Score'].map({'Good':1,
                                                               'Poor':0,
                                                               'Standard':2})
credit_data['Credit_Mix'] = credit_data['Credit_Mix'].map({'Good':1,
                                                            'Bad':0,
                                                         'Standard':2})

x = np.array(credit_data[['Annual_Income', 'Monthly_Inhand_Salary', 'Num_Bank_Accounts',
                          'Num_Credit_Card', 'Interest_Rate', 'Num_of_Loan', 'Delay_from_due_date',
                          'Num_of_Delayed_Payment', 'Credit_Mix', 'Outstanding_Debt',
                          'Credit_Utilization_Ratio', 'Credit_History_Age', 'Total_EMI_per_month',
                          'Amount_invested_monthly', 'Monthly_Balance']])

y = np.array(credit_data['Credit_Score'])

x_train_rf, x_test_rf, y_train_rf, y_test_rf = train_test_split(x, y, test_size=0.3, random_state=42)

Standardscaler = StandardScaler()
x_train_rf = Standardscaler.fit_transform(x_train_rf)
x_test_rf = Standardscaler.transform(x_test_rf)

rf = RandomForestClassifier(n_estimators=10)
rf.fit(x_train_rf,y_train_rf)

y_pred_rf = rf.predict(x_test_rf)

accuracyscore_rf = accuracy_score(y_test_rf,y_pred_rf)
classificationreport_rf = classification_report(y_test_rf, y_pred_rf)
confusionmatrix_rf = confusion_matrix(y_test_rf,y_pred_rf)

print('accuracyscore_rf: {:.4f}'.format(accuracyscore_rf))
print('classificationreport_rf: \n',classificationreport_rf)
print('confusionmatrix_rf: \n',confusionmatrix_rf)

```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt


n_trees = len(rf.estimators_)
for i in range(n_trees):
    plt.figure(figsize=(10,10))
    plot_tree(rf.estimator_[i], feature_names=['Annual_Income', 'Monthly_Inhand_Salary', 'Num_Bank_Accounts',
                          'Num_Credit_Card', 'Interest_Rate', 'Num_of_Loan', 'Delay_from_due_date',
                          'Num_of_Delayed_Payment', 'Credit_Mix', 'Outstanding_Debt',
                          'Credit_Utilization_Ratio', 'Credit_History_Age', 'Total_EMI_per_month',
                          'Amount_invested_monthly', 'Monthly_Balance'],
                                               class_names=['Poor', 'Good', 'Standard'], filled=True, rounded=True)

plt.show()
```
